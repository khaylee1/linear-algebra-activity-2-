```python
def eagle_kinematics(displacement, trajectory):
    req_shape = 4
    velocity = np.zeros((req_shape-1,))
    acceleration = np.zeros((req_shape-2,))
    trajectory_vect = np.array([trajectory**3, trajectory**2, trajectory, 1])
    if displacement.shape == (req_shape,):
        vector = np.array([3*displacement[0],2*displacement[1], displacement[2]])
        acceleration_trajectory = np.array([2*velocity[0],velocity[1]])
        displacement_trajectory = np.sum(np.multiply(displacement, trajectory_vect))
        velocity_trajectory = np.sum(np.multiply(vector, trajectory_vect[1:]))
        acceleration_trajectory = np.sum(np.multiply(acceleration, trajectory_vect[2:]))
        
        plt.figure(figsize=(10,7))
        plt.title('Philippine Eagle Flight Kinematics')
        plt.xlim(-30, 30)
        plt.ylim(-30, 30)
        plt.xlabel('Latitudinal Distance')
        plt.ylabel('Longitudinal Distance')
        plt.grid()
        n = 2 
        
        #plt.quiver(0,0, vector[0],vector[1],
         #      angles='xy', scale_units='xy',scale=1, color ='blue', 
          #     label='velocity: {:.2f}m/s.'.format(np.linalg.norm(vector)))
        plt.quiver(np.add(0,0), np.add(0,0),
               vector[0], vector[1], angles='xy', scale_units='xy',scale=1, color='green',
              label='Trajectory 3: {:.2f}m.'.format(np.linalg.norm(vector)))
        plt.quiver(0,0, acceleration[0], acceleration[1], 
               angles='xy', scale_units='xy',scale=1, color='red', width= 0.05, 
               label='acceleration: {:.2f}m.'.format(np.linalg.norm(acceleration)))
        #plt.quiver(0,0, trajectory_vect[0], trajectory_vect[1],trajectory_vect[2], 
         #      angles='xy', scale_units='xy',scale=1, color='green', 
          #     label='displacement3: {:.2f}m.'.format(np.linalg.norm(trajectory_vect)))
        
        plt.quiver(np.add(0,0), np.add(0,0),
               trajectory_vect[0], trajectory_vect[1], angles='xy', scale_units='xy',scale=1, color='blue',
              label='Trajectory 3: {:.2f}m.'.format(np.linalg.norm(trajectory_vect)))
        plt.legend()
        
        
    else:
        print(f'Input displacement vector is not valid. Make sure that the vector shape is ({req_shape},)')
    
    return displacement_trajectory, velocity_trajectory, acceleration_trajectory
```
